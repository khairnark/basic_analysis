### Visualization for PageSense Accuracy Status are as following:

```python
PageSenseFolderCount = df.groupby('PageSense Accuracy Status')['Folder ID'].count().reset_index()
PageSenseFolderCount
```

```python
ax = sns.countplot(data = df, x = 'PageSense Accuracy Status')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Visualization for SDRS Accuracy Status are as following:

```python
sdrsFolderCount = df.groupby('SDRS Accuracy Status')['Folder ID'].count().reset_index()
sdrsFolderCount
```

```python
ax = sns.countplot(data = df, x = 'SDRS Accuracy Status')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Visualization for SDRS Confidence Band are as following:

```python
bandFolderCount = df.groupby('SDRS Confidence Band')['Folder ID'].count().reset_index()
bandFolderCount
```

```python
ax = sns.countplot(data = df, x = 'SDRS Confidence Band')

sns.set(rc={'figure.figsize':(11,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Visualization for count of total folder per engine

```python
#count of total folder per engine
FolderCount = df['Folder ID'].groupby(df['Selected Engine']).count().reset_index()
FolderCount
```

```python
ax = sns.countplot(data = df, x = 'Selected Engine')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```
